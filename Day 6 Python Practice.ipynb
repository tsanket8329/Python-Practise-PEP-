```python
#Super is the keyword to inherit the constructorof the parent class by the child
class Parent:
    def __init__(self):
        print("this is parent constructor")

class Child(Parent):
  def __init__(self):
    super().__init__()
    print("this is child constructor")
obj = Child()
```

```python
class Parent:
    def shape(self):
        print("It is a nice shape")

class Child(Parent):
    def shape(self):
        super().shape()      # calls Parent's shape()
        print("It is a circle")

obj = Child()
obj.shape()
```

```python
#Create a parent class Account and it should have protected variable _balance then constructor should set the balance. method show_balance() should print balance
#child class saving account which will inherit from account and create a method add_moeny(amount) and add amount to balance and show updated balance

class Account:
    def __init__(self, balance):
        self._balance = balance
    def show_balance(self):
        print("Balance:", self._balance)
class SavingAccount(Account):
    def add_money(self, amount):
        self._balance += amount
        print("Money added:", amount)
        self.show_balance()
s1 = SavingAccount(5000)
s1.show_balance()
s1.add_money(2000)
```

```python
#Diamond Problem
class A:
    def message(self):
        print("this is message from class A")

class B(A):
    def message(self):
        print("this is message from class B")

class C(A):
    def message(self):
        print("this is message from class C")

class D(B, C):
  def message(self):
        print("this is message from class D")


obj = D()
obj.message()
```

```python
#Polymorphism is one name many forms same method name different behaviour which depends upon object
class Human:
  def speak(self):
    print("Human is speaking")
class Baby:
  def speak(self):
    print("baby is crying")
class Dog:
  def speak(self):
    print("Dog is barking")
#polymorphism in action
h = Human()
b = Baby()
d=Dog()
h.speak()
b.speak()
d.speak()
```

```python
#create a class person with method role( create 2 class student and teacher student prints i m a student and teacher print i m a teacher
#create another class assistant that inherit frtom both studeny and teacher create an obj of assistant call role( observer which method is called
# Base class
class Person:
    def role(self):
        print("I am a person")

# First child class
class Student(Person):
    def role(self):
        print("I am a student")

# Second child class
class Teacher(Person):
    def role(self):
        print("I am a teacher")

# Multiple inheritance
class Assistant(Student, Teacher):
    pass

# Create object of Assistant
obj = Assistant()
obj.role()
```

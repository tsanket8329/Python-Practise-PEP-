```python
# Dictionary to store contacts (name : phone number)
contact = {}

# Infinite loop to keep the program running until user exits
while True:
    print("\n---- CONTACT BOOK ----")
    print("1. Add contact")
    print("2. View contacts")
    print("3. Search contact")
    print("4. Delete contact")
    print("5. Exit")

    # Taking user choice
    choice = input("Enter your choice here: ")

    # OPTION 1: Add a new contact
    if choice == "1":
        name = input("Enter name: ")
        phone = input("Enter your phone number: ")
        contact[name] = phone      # Store contact in dictionary
        print("Contact added successfully")

    # OPTION 2: View all saved contacts
    elif choice == "2":
        if contact:                # Check if dictionary is not empty
            print("\nSaved contacts:")
            for name, phone in contact.items():
                print(name, ":", phone)
        else:
            print("No contact found")

    # OPTION 3: Search for a contact
    elif choice == "3":
        name = input("Enter name to search: ")
        if name in contact:
            print("Phone number:", contact[name])
        else:
            print("Contact not found")

    # OPTION 4: Delete a contact
    elif choice == "4":
        name = input("Enter the name you want to delete: ")
        if name in contact:
            del contact[name]      # Remove contact from dictionary
            print("Contact deleted")
        else:
            print("Contact not found")

    # OPTION 5: Exit the program
    elif choice == "5":
        print("Thank you for using contact book")
        break                      # Exit the while loop

    # If user enters invalid option
    else:
        print("Invalid choice. Please try again.")
```

```python
# Dictionary to store student details
# roll_no : {"name": name, "course": course}
Student = {}

# Infinite loop to keep the program running
while True:
    print("\n---- STUDENT DETAILS ----")
    print("1. Add student")
    print("2. View all students")
    print("3. Search student")
    print("4. Delete student")
    print("5. Exit")

    # User choice
    choice = input("Enter your choice here: ")

    # OPTION 1: Add student
    if choice == "1":
        roll = input("Enter Roll Number: ")
        name = input("Enter Student Name: ")
        course = input("Enter Course Name: ")

        # Store student details as a nested dictionary
        Student[roll] = {
            "Name": name,
            "Course": course
        }

        print("Student added successfully")

    # OPTION 2: View all students
    elif choice == "2":
        if Student:
            print("\nStudent Records:")
            for roll, details in Student.items():
                print("----------------------")
                print("Roll No :", roll)
                print("Name    :", details["Name"])
                print("Course  :", details["Course"])
        else:
            print("No student record found")

    # OPTION 3: Search student
    elif choice == "3":
        roll = input("Enter Roll Number to search: ")
        if roll in Student:
            print("\nStudent Found:")
            print("Roll No :", roll)
            print("Name    :", Student[roll]["Name"])
            print("Course  :", Student[roll]["Course"])
        else:
            print("Student not found")

    # OPTION 4: Delete student
    elif choice == "4":
        roll = input("Enter Roll Number to delete: ")
        if roll in Student:
            del Student[roll]
            print("Student deleted successfully")
        else:
            print("Student not found")

    # OPTION 5: Exit program
    elif choice == "5":
        print("Thank you for using Student Management System")
        break

    # Invalid option
    else:
        print("Invalid choice. Please try again.")
```

```python
# String manipulation examples

# Creating a string
s = "python programming"

# Length of the string
print(len(s))              # Output: total number of characters

# Last character of the string
print(s[-1])               # Output: 'g'

# Slicing: first 4 characters
print(s[:4])               # Output: 'pyth'

# Slicing from index 7 to 17
print(s[7:18])              # Output: 'programming'

# Convert string to lowercase
print(s.lower())           # Output: 'python programming'

# Convert string to uppercase
print(s.upper())           # Output: 'PYTHON PROGRAMMING'

# Remove leading and trailing spaces (if any)
print(s.strip())           # Output: 'python programming'

# Replace a word in the string
print(s.replace("Python", "java"))
# Note: This will not change anything because 'Python' != 'python'

# Find the first occurrence of character 'o'
print(s.find("o"))         # Output: index position of 'o'

# Split the string into a list of words
print(s.split())           # Output: ['python', 'programming']

# Check if all characters are alphabets
a = "abc"
print(a.isalpha())         # Output: True

# Check if all characters are digits
print(a.isdigit())         # Output: False

# String containing only digits
b = "123"
print(b.isdigit())         # Output: True

# Count occurrences of character 'p'
print(s.count("p"))        # Output: number of times 'p' appears

# Print original string
print(s)

# Check if string starts with 'py'
print(s.startswith("py"))  # Output: True

# Check if string ends with 'ing'
print(s.endswith("ing"))   # Output: True
```

```python
# String
s = "Programming"

# Two pointers
start = 0
end = len(s) - 1

# Loop until pointers cross
while start <= end:
    print("Start:", s[start], "| End:", s[end])
    start += 1
    end -= 1
```

```python
# String
s = "madam"

# Two pointers
start = 0
end = len(s) - 1

# Assume palindrome
is_palindrome = True

while start < end:
    if s[start] != s[end]:
        is_palindrome = False
        break
    start += 1
    end -= 1

# Result
if is_palindrome:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
A string in Python is simply a sequence of characters that represents text, like "Hello World". It is of type str and is immutable, which means you cannot change it directly. On the other hand, the split() method is used to break a string into smaller parts, usually words, and returns them as a list. For example, "Hello World".split() will give ['Hello', 'World']. This allows you to work with each word individually. So, in easy terms, a string is the full text, and split() is a way to divide that text into separate pieces that you can process or analyze.
```

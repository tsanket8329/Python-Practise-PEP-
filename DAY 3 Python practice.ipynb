```python
ZeroDivisionError - When u divide any no. with 0 --> print(10/0) #dividing by zero
ValueError - When u give wrong value to a function --> num=int("abc")#wrong value
TypeError - When u perform an operation between different data types --> print(5+"abc")#wrong data value
IndexError - When u access the wrong index --> list[10,20,30]..print(list[5])#wrong index
KeyError - When u access a missing key in a dictionary -- students={"name":"Rahul","age":20}  print(student["marks"])# missing key
FileNotFoundError - When python not able to find file in your pc #file not found
```

```python
#Move all zeros to the end
nums=[0,1,0,3,12]
result=[] #we will store non-zero elements
zero_count=0
for n in nums:
  if n==0:
    zero_count+=1
  else:
    result.append(n)
for i in range(zero_count):
  result.append(0)
print(result)
```

```python
text = "muskan"
left=0
right=len(text)-1
isPalindrome = True
while left<right:
  if text[left]!=text[right]:
    isPalindrome = False
    break
  left+=1
  right+=1
print(isPalindrome)
```

```python
num=[1,2,3,2]
target=5
count=0
for i in range(len(num)):
  total = 0
  for i in range(i,len(num)):
    total+=num[i]
    if total==target:
      count+=1
print(count)
```

```python
#brute force approach it has nested for loops time complexity is more or worst
num=[2,7,11,15]
target=9
count=0
for i in range(len(num)):
  for j in range(i+1,len(num)):
    #check if sum matches the target
     if num[i]+num[j]==target:
      print(num[i],num[j])
```

```python
#we are using two pointer approach to reduce loops and time complexity
num=[2,7,11,15]
target=9
left=0
right=len(num)-1
while left<right:
  current_sum = num[left]+num[right]
  if current_sum == target:
    print(num[left]+num[right])
    break
  elif current_sum <target:
    left+=1
  else:
    right-=1
```

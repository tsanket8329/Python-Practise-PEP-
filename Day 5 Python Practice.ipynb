```python
#Encapsulation
class Student:
    def __init__(self, marks):
        self.__marks = marks   # private variable

    def get_marks(self):
        return self.__marks
    def set_marks(self,new_marks):
      self.__marks=new_marks

s1 = Student(100)
s1.set_marks(90)
print(s1.get_marks())
```

```python
#create a class in which balance will be hidden and access it using get_balance method
class Bank:
    def __init__(self, balance):
        self.__balance = balance   # private variable

    def get_balance(self):
        return self.__balance      # getter method
    def set_balance(self, new_balance):
      self.__balance = new_balance

b1 = Bank(100)
b1.set_balance(80)
print(b1.get_balance())
```

```python
class Account:
    def __init__(self, balance):
        self.__balance = balance   # private variable

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Amount deposited successfully")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawal successful")
        else:
            print("Insufficient balance")

a1 = Account(1000)

print("Balance:", a1.get_balance())
a1.deposit(500)
print("Balance after deposit:", a1.get_balance())
a1.withdraw(300)
print("Balance after withdrawal:", a1.get_balance())
```
